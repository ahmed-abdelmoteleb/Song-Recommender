```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import difflib
```

## Storing the URL

```python
url = "https://www.billboard.com/charts/hot-100"
```

## Getting the html code of the web page

```python
response = requests.get(url)
response.status_code # 200 status code means OK!
```

## Parsing the html code

```python
soup = BeautifulSoup(response.content, "html.parser")
```

## Retrieving the desired info from the Soup.

```python
##charts > div > div.chart-list__wrapper > div > ol > li:nth-child(1)
soup.select("ol")
```

```python
song = [ i.get_text().lower() for i in soup.select(".chart-element__information__song")]
   
```
